# External Driver Coupling Analysis

This notebook analyzes the relationship between Humboldt Current Marine Heatwaves and macro-basin drivers, notably the El Niño-Southern Oscillation (ENSO).

We use the Oceanic Niño Index (ONI) and detrended SST anomalies decomposed via PCA to track coastally trapped Kelvin waves.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook", font_scale=1.2)
```

## 1. Load PCA and ONI Data

```python
# Load merged PC1 and ONI dataset
df_pc1 = pd.read_csv('../data/processed/pc1_oni_merged.csv', parse_dates=['time'])
df_pc1.set_index('time', inplace=True)

print("Correlation between EOF1 (PC1) and ONI:")
print(df_pc1[['PC1', 'ONI']].corr())
```

## 2. Visualize PC1 vs ONI

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(df_pc1.index, df_pc1['PC1'] / df_pc1['PC1'].std(), label='Normalized PC1 (Humboldt SST Variance)', color='firebrick')
ax.plot(df_pc1.index, df_pc1['ONI'] / df_pc1['ONI'].std(), label='Normalized ONI (Niño 3.4)', color='dodgerblue', alpha=0.7)
ax.set_title("Humboldt PC1 vs Oceanic Niño Index (1982-2024)")
ax.set_ylabel("Standard Deviations")
ax.legend()
plt.show()
```

## 3. Lag Correlation Analysis

```python
df_lag = pd.read_csv('../data/processed/oni_mhw_lag_correlation.csv')

plt.figure(figsize=(10, 5))
plt.bar(df_lag['lag_months'], df_lag['correlation'], color='coral')
plt.axvline(0, color='black', linestyle='--')
plt.title('Lag Correlation: ONI vs Humboldt MHW Intensity')
plt.xlabel('Lag (Months)')
plt.ylabel('Pearson Correlation')
plt.show()

print("Peak correlation lag:")
print(df_lag.loc[df_lag['correlation'].idxmax()])
```
